```python
import matplotlib.pyplot as plt

from skimage.morphology import convex_hull_image
from skimage import data, img_as_float
from skimage.util import invert
```

## Invert the original Image

* Import a built-in <b>binary image</b> of <b>horse</b> from <b>data module</b>
* <b>Invert</b> the <b>horse_Image</b> and store Inverted data into a new object

```python
horse_image = data.horse()   

horse_inverted = invert(data.horse()) 
```

```python
fig, axes = plt.subplots(1, 2, 
                         figsize=(12, 8), 
                         sharex=True, sharey=True)  

ax = axes.ravel() 

ax[0].set_title('Original')
ax[0].imshow(horse_image, cmap='gray')

ax[1].set_title('Inverted')
ax[1].imshow(horse_inverted, cmap='gray')

fig.tight_layout()  
plt.show()  
```

## Find the Convex Hull of an Image

It will find the <b>smallest convex polygon</b> which will surround all white pixels in inverted horse image

```python
convexhull_horse = convex_hull_image(horse_inverted)  
```

### Plot Convex Hull of an Image

* Divide the Figure into <b>1 row, 2 columns</b>
* Plot <b>Inverted_horse_image</b> and <b>Original horse Image</b> with <b>Convex Hull</b> 


```python
fig, axes = plt.subplots(1, 2, 
                         figsize=(12, 12), 
                         sharex=True, sharey=True) 
ax = axes.ravel()

ax[0].set_title('Inverted Image')
ax[0].imshow(horse_inverted,
             cmap='gray', 
             interpolation='nearest')  

ax[1].set_title('Convex Hull')
ax[1].imshow(convexhull_horse, 
             cmap='gray', 
             interpolation='nearest')  

plt.tight_layout()  
plt.show()
```

## Identify the difference between Inverted and Transformed Image

* Copy the image with convex hull and convert it in float to new object 
* Fill the inverted horse image to differentiate convex hull and the horse object


```python
convexhull_diff = img_as_float(convexhull_horse.copy())  

convexhull_diff[horse_inverted] = 2
```

* Plot the difference between Inverted image and Transformed image

```python
fig, ax = plt.subplots(figsize=(12, 8)) 

ax.imshow(convexhull_diff, 
          cmap='gray', 
          interpolation='nearest')

ax.set_title('Difference between Inverted and Convex Hull')

plt.show()
```
